```python
''''
Creo una función que excluye todos los elementos de la lista 'no_paises' de la lista 'paises'.
'''

def excluir_no_pais(paises, no_paises):
  return [pais for pais in paises if pais not in no_paises]

paises_filtrados = excluir_no_pais(paises, no_paises)

print(paises_filtrados)
```

```python
'''
Top 3 paises productores de CO2 en 2022
'''

# Filtrar el dataframe para el año 2022
df_2022 = df[df['ano'] == 2022]

# Filtrar el dataframe para incluir solo países
df_2022 = df_2022[df_2022['pais'].isin(paises_filtrados)]

# Agrupar por país y sumar las emisiones de CO2
df_agrupado = df_2022.groupby('pais')['emisiones_co2'].sum().sort_values(ascending=False)

# Obtener los 3 países con mayor producción de CO2
top_3_paises = df_agrupado.head(3)

print(top_3_paises)
# Crear un gráfico de barras
top_3_paises.plot(kind='bar')
plt.title('Top 3 países con mayor producción de CO2 en 2022')
plt.xlabel('País')
plt.ylabel('Emisiones de CO2')
plt.show()
```

**Interpretación gráfico:** En el gráfico se muestra los 3 principales países productres de emisiones de CO2. En donde China (nro 1) produce más de el doble que USA (nro 2) y casi 3 veces lo que produce India (nro 3)

```python
'''
Top 3 paises productores de Gas en 2022
'''

# Filtrar el dataframe para el año 2022
df_2022 = df[df['ano'] == 2022]

# Filtrar el dataframe para incluir solo países
df_2022 = df_2022[df_2022['pais'].isin(paises_filtrados)]

# Agrupar por país y sumar la produccion de Gas
df_agrupado = df_2022.groupby('pais')['prod_gas'].sum().sort_values(ascending=False)

# Obtener los 3 países con mayor producción de Gas
top_3_paises = df_agrupado.head(3)

print(top_3_paises)
# Crear un gráfico de barras
top_3_paises.plot(kind='bar')
plt.title('Top 3 países con mayor producción de gas en 2022')
plt.xlabel('País')
plt.ylabel('Produccion de Gas')
plt.show()
```
